```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi
```

```python
song = 'SpotifyAudioFeaturesApril2019.csv'
song1 = 'SpotifyAudioFeaturesNov2018.csv'
df = pd.read_csv(song)
df1 = pd.read_csv(song1)
```

```python
print(df.shape)
df.head()
```

```python
print(df1.shape)
df1.head()
```

```python
df.describe()
```

```python
df1.describe()
```

```python
#merge the DataFrames
full = pd.concat([df, df1], ignore_index=True)
print(full.shape)
full.head()
```

```python
full.duplicated().value_counts()
```

```python
#No duplicate
full = full.drop_duplicates('track_id', keep='first')
```

```python
full.duplicated().value_counts()
```

```python
full['track_index'] = full.index
full.track_index
```

```python
area_dict = dict(zip(full['track_name'], full['track_index']))
area_dict
```

```python
#class
input_string = input()
input_string2 = input()
dict_lookup_0 = area_dict[input_string]
dict_lookup_1 = area_dict[input_string2]

plt.style.use('fivethirtyeight')
# number of variable
categories=list(full)[1:]
N = len(categories)
# What will be the angle of each axis in the plot? (we divide the plot / number of variable)
angles = [n / float(N) * 2 * pi for n in range(N)]
angles += angles[:1]
# Initialise the spider plot
ax = plt.subplot(111, polar=True)
# If you want the first axis to be on top:
ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)
# Draw one axe per variable + add labels labels yet
plt.xticks(angles[:-1], categories, size=10)
# Draw ylabels
ax.set_rlabel_position(0)
plt.yticks([10,20,30], ["10","20","30"], color="grey", size=5)
plt.ylim(0,40)
# ------- PART 2: Add plots
# Plot each individual = each line of the data
# I don't do a loop, because plotting more than 3 groups makes the chart unreadable
# Ind1
values=full.loc[dict_lookup_0].drop('acousticness').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=input_string)
ax.fill(angles, values, 'b', alpha=0.1)
# Ind2
values=full.loc[dict_lookup_1].drop('acousticness').values.flatten().tolist()
values += values[:1]
ax.plot(angles, values, linewidth=1, linestyle='solid', label=input_string2)
ax.fill(angles, values, 'r', alpha=0.1)
# Add legend
plt.legend(loc='upper right', bbox_to_anchor=(0.01, 0.01))
```
